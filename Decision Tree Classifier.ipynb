```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
import pandas as pd

# Load the dataset from the local file system
file_path = "C:/Users/hp/OneDrive/Desktop/VCU/Sparks Foundation/Iris.csv"  # Replace with your file path

# Load the CSV file into a DataFrame
data = pd.read_csv(file_path)
print("Data loaded successfully!")
print(data.head())
```

```python
# Inspect the dataset to understand the structure and features
print(data.info())
print(data.describe())
```

```python
# Selecting features and the target variable
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
target = 'Species'

# Prepare the feature matrix (X) and the target vector (y)
X = data[features]
y = data[target]

# Handling missing values (if any)
X.fillna(0, inplace=True)
y.fillna(0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Plot the decision tree
plt.figure(figsize=(20,10))
tree.plot_tree(clf, filled=True, feature_names=features, class_names=clf.classes_)
plt.show()
```
